## Import packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Visualisations

## Scatter plot
Does the LLM agree with humans?

```python
comparison_df = human_comment_stats_df.merge(
    llm_comment_stats_df,
    on="comment_id",
    suffixes=("_human", "_llm")
)

plt.figure()

sns.scatterplot(
    data=comparison_df,
    x="mean_irony_human",
    y="mean_irony_llm"
)

# diagonal line (perfect agreement)
plt.plot([1,5], [1,5], linestyle="--")

plt.xlabel("Human Mean Irony")
plt.ylabel("LLM Mean Irony")
plt.title("Human vs LLM Irony Ratings")

plt.savefig("figures/comparison/human_vs_llm_irony.png", dpi=300, bbox_inches="tight")
plt.show()
```
